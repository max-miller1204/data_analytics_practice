```python
import pandas as pd
```

```python
df = pd.read_csv('PlayerIndex_nba_stats.csv')
```

```python
# Filter for career stats
career_stats = df[df['STATS_TIMEFRAME'] == 'Season']

# Filter for players with career points per game over 25
high_scorers = career_stats[career_stats['PTS'] > 30]
```

```python
high_scorers
```

```python
r15_scorers = df[df['DRAFT_ROUND'] == 15]

high_scorers = r15_scorers[r15_scorers['PTS'] > 11]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ai_dev_productivity.csv')
```

```python
# Display basic info and first few rows
info = df.info()
head = df.head()

# Summary statistics
summary_stats = df.describe()
```

```python
# Plot distributions of key columns
plt.figure(figsize=(12, 8))
sns.histplot(df['task_success'], kde=False, bins=2)
plt.title('Distribution of Task Success')
plt.show()

plt.figure(figsize=(12, 8))
sns.histplot(df['cognitive_load'], kde=True)
plt.title('Distribution of Cognitive Load')
plt.show()

(info, head, summary_stats)
```
